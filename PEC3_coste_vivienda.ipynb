## 3. Transformación de los datos

```python
# realizamos un filtrado de los datos, para obtener únicamente el precio total en euros por m2 construido
precio_medio_2019_trimestres = df_precio_2019[df_precio_2019['Preu_mitja_habitatge'].str.contains("Total. Euros/m2 construït", regex=False)]
precio_medio_2018_trimestres = df_precio_2018[df_precio_2018['Preu_mitja_habitatge'].str.contains("Total. Euros/m2 construït", regex=False)]
precio_medio_2017_trimestres = df_precio_2017[df_precio_2017['Preu_mitja_habitatge'].str.contains("Total. Euros/m2 construït", regex=False)]
precio_medio_2016_trimestres = df_precio_2016[df_precio_2016['Preu_mitja_habitatge'].str.contains("Total. Euros/m2 construït", regex=False)]
precio_medio_2015_trimestres = df_precio_2015[df_precio_2015['Preu_mitja_habitatge'].str.contains("Total. Euros/m2 construït", regex=False)]
```

```python
# visualizamos los resultados
precio_medio_2015_trimestres
```

```python
# identificamos el tipo de variable en cada columna
precio_medio_2015_trimestres.dtypes
```

```python
# eliminamos las columnas que no necesitamos
precio_medio_2019_trimestres = precio_medio_2019_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre', 'Preu_mitja_habitatge'], axis=1)
precio_medio_2018_trimestres = precio_medio_2018_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre', 'Preu_mitja_habitatge'], axis=1)
precio_medio_2017_trimestres = precio_medio_2017_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre', 'Preu_mitja_habitatge'], axis=1)
precio_medio_2016_trimestres = precio_medio_2016_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre', 'Preu_mitja_habitatge'], axis=1)
precio_medio_2015_trimestres = precio_medio_2015_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre', 'Preu_mitja_habitatge'], axis=1)
```

```python
# visualizamos los resultados
precio_medio_2015_trimestres
```

```python
# Se identifican valores erróneos o ausentes en la columna referente al valor de la vivienda, por lo que procedemos a imputar estos valores con el valor de la media del precio de venta
precio_medio_2019_trimestres['Valor'] = precio_medio_2019_trimestres['Valor'].replace('--', np.nan, regex = False)
precio_medio_2018_trimestres['Valor'] = precio_medio_2018_trimestres['Valor'].replace('--', np.nan, regex = False)
precio_medio_2017_trimestres['Valor'] = precio_medio_2017_trimestres['Valor'].replace('--', np.nan, regex = False)
precio_medio_2016_trimestres['Valor'] = precio_medio_2016_trimestres['Valor'].replace('--', np.nan, regex = False)
precio_medio_2015_trimestres['Valor'] = precio_medio_2015_trimestres['Valor'].replace('--', np.nan, regex = False)
```

```python
precio_medio_2019_trimestres['Valor'] = precio_medio_2019_trimestres['Valor'].astype(np.float64)
precio_medio_2018_trimestres['Valor'] = precio_medio_2018_trimestres['Valor'].astype(np.float64)
precio_medio_2017_trimestres['Valor'] = precio_medio_2017_trimestres['Valor'].astype(np.float64)
precio_medio_2016_trimestres['Valor'] = precio_medio_2016_trimestres['Valor'].astype(np.float64)
precio_medio_2015_trimestres['Valor'] = precio_medio_2015_trimestres['Valor'].astype(np.float64)
```

```python
precio_medio_2019_trimestres['Valor'] = precio_medio_2019_trimestres['Valor'].fillna((precio_medio_2019_trimestres['Valor'].mean()))
precio_medio_2018_trimestres['Valor'] = precio_medio_2018_trimestres['Valor'].fillna((precio_medio_2018_trimestres['Valor'].mean()))
precio_medio_2017_trimestres['Valor'] = precio_medio_2017_trimestres['Valor'].fillna((precio_medio_2017_trimestres['Valor'].mean()))
precio_medio_2016_trimestres['Valor'] = precio_medio_2016_trimestres['Valor'].fillna((precio_medio_2016_trimestres['Valor'].mean()))
precio_medio_2015_trimestres['Valor'] = precio_medio_2015_trimestres['Valor'].fillna((precio_medio_2015_trimestres['Valor'].mean()))
```

```python
# Como los datos vienen por trimestres, realizamos la agrupación por barrio y calculamos el precio medio de venta por barrio
grouped_precio_medio_2019 = precio_medio_2019_trimestres.groupby(['Codi_Barri']).mean().reset_index()
grouped_precio_medio_2018 = precio_medio_2018_trimestres.groupby(['Codi_Barri']).mean().reset_index()
grouped_precio_medio_2017 = precio_medio_2017_trimestres.groupby(['Codi_Barri']).mean().reset_index()
grouped_precio_medio_2016 = precio_medio_2016_trimestres.groupby(['Codi_Barri']).mean().reset_index()
grouped_precio_medio_2015 = precio_medio_2015_trimestres.groupby(['Codi_Barri']).mean().reset_index()
```

```python
# visualizamos los resultados
grouped_precio_medio_2015
```

```python
# se renombra la columna valor para ser más descriptiva
grouped_precio_medio_2019.columns = grouped_precio_medio_2019.columns.str.replace('Valor', 'precio_vivienda_Euros/m2')
grouped_precio_medio_2018.columns = grouped_precio_medio_2018.columns.str.replace('Valor', 'precio_vivienda_Euros/m2')
grouped_precio_medio_2017.columns = grouped_precio_medio_2017.columns.str.replace('Valor', 'precio_vivienda_Euros/m2')
grouped_precio_medio_2016.columns = grouped_precio_medio_2016.columns.str.replace('Valor', 'precio_vivienda_Euros/m2')
grouped_precio_medio_2015.columns = grouped_precio_medio_2015.columns.str.replace('Valor', 'precio_vivienda_Euros/m2')
```

```python
grouped_precio_medio_2019.isnull().any()
```

## 4. Concatenación de los dataframes

```python
# relizamos la concatenación de los datasets en un unico dataset
```

```python
union_precio_medio = pd.concat([grouped_precio_medio_2019, grouped_precio_medio_2018, grouped_precio_medio_2017, \
                             grouped_precio_medio_2016, grouped_precio_medio_2015], ignore_index=True)
```
