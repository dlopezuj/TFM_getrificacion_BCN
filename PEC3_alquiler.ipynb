```python
# se realiza la lectura de los diversos archivos csv y se guardan en cinco datasets (uno para cada año)
df_alquiler_2019=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_economica/coste alquiler/2019_lloguer_preu_trim.csv')
df_alquiler_2018=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_economica/coste alquiler/2018_lloguer_preu_trim.csv')
df_alquiler_2017=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_economica/coste alquiler/2017_lloguer_preu_trim.csv')
df_alquiler_2016=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_economica/coste alquiler/2016_lloguer_preu_trim.csv')
df_alquiler_2015=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_economica/coste alquiler/2015_lloguer_preu_trim.csv')
```

```python
# filtramos los datos para crear un nuevo dataframe únicamente con los resultados de alquiler medio por superficie
alquiler_medio_sup_mesual_2019_trimestres = df_alquiler_2019[df_alquiler_2019['Lloguer_mitja'].str.contains('Lloguer mitjà per superfície (Euros/m2 mes)', regex=False)]
alquiler_medio_sup_mesual_2018_trimestres = df_alquiler_2018[df_alquiler_2018['Lloguer_mitja'].str.contains('Lloguer mitjà per superfície (Euros/m2 mes)', regex=False)]
alquiler_medio_sup_mesual_2017_trimestres = df_alquiler_2017[df_alquiler_2017['Lloguer_mitja'].str.contains('Lloguer mitjà per superfície (Euros/m2 mes)', regex=False)]
alquiler_medio_sup_mesual_2016_trimestres = df_alquiler_2016[df_alquiler_2016['Lloguer_mitja'].str.contains('Lloguer mitjà per superfície (Euros/m2 mes)', regex=False)]
alquiler_medio_sup_mesual_2015_trimestres = df_alquiler_2015[df_alquiler_2015['Lloguer_mitja'].str.contains('Lloguer mitjà per superfície (Euros/m2 mes)', regex=False)]

```

## 3. Transformación de los datos

```python
# Se decide reemplazar los valores null, por la media de los precios del alquiler
alquiler_medio_sup_mesual_2019_trimestres['Preu'] = alquiler_medio_sup_mesual_2019_trimestres['Preu'].fillna((alquiler_medio_sup_mesual_2019_trimestres['Preu'].mean()))
alquiler_medio_sup_mesual_2018_trimestres['Preu'] = alquiler_medio_sup_mesual_2018_trimestres['Preu'].fillna((alquiler_medio_sup_mesual_2018_trimestres['Preu'].mean()))
alquiler_medio_sup_mesual_2017_trimestres['Preu'] = alquiler_medio_sup_mesual_2017_trimestres['Preu'].fillna((alquiler_medio_sup_mesual_2017_trimestres['Preu'].mean()))
alquiler_medio_sup_mesual_2016_trimestres['Preu'] = alquiler_medio_sup_mesual_2016_trimestres['Preu'].fillna((alquiler_medio_sup_mesual_2016_trimestres['Preu'].mean()))
alquiler_medio_sup_mesual_2015_trimestres['Preu'] = alquiler_medio_sup_mesual_2015_trimestres['Preu'].fillna((alquiler_medio_sup_mesual_2015_trimestres['Preu'].mean()))
```

```python
alquiler_medio_sup_mesual_2015_trimestres.isnull().sum()
```

```python
# se eliminan las columnas que no son necesarias
alquiler_medio_sup_mesual_2019 = alquiler_medio_sup_mesual_2019_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre'], axis=1)
alquiler_medio_sup_mesual_2018 = alquiler_medio_sup_mesual_2018_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre'], axis=1)
alquiler_medio_sup_mesual_2017 = alquiler_medio_sup_mesual_2017_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre'], axis=1)
alquiler_medio_sup_mesual_2016 = alquiler_medio_sup_mesual_2016_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre'], axis=1)
alquiler_medio_sup_mesual_2015 = alquiler_medio_sup_mesual_2015_trimestres.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Trimestre'], axis=1)
```

```python
# realizamos el calculo de la media del alquiler por barrio
grouped_alquiler_medio_sup_mesual_2019 = alquiler_medio_sup_mesual_2019.groupby(['Codi_Barri']).mean().reset_index()
grouped_alquiler_medio_sup_mesual_2018 = alquiler_medio_sup_mesual_2018.groupby(['Codi_Barri']).mean().reset_index()
grouped_alquiler_medio_sup_mesual_2017 = alquiler_medio_sup_mesual_2017.groupby(['Codi_Barri']).mean().reset_index()
grouped_alquiler_medio_sup_mesual_2016 = alquiler_medio_sup_mesual_2016.groupby(['Codi_Barri']).mean().reset_index()
grouped_alquiler_medio_sup_mesual_2015 = alquiler_medio_sup_mesual_2015.groupby(['Codi_Barri']).mean().reset_index()
```

```python
# visualizamos los resultados obtenidos
grouped_alquiler_medio_sup_mesual_2016
```

```python
# renombramos la columna preu
grouped_alquiler_medio_sup_mesual_2019.columns = grouped_alquiler_medio_sup_mesual_2019.columns.str.replace('Preu', 'Alquiler_medio(Euros/m2 mes)')
grouped_alquiler_medio_sup_mesual_2018.columns = grouped_alquiler_medio_sup_mesual_2018.columns.str.replace('Preu', 'Alquiler_medio(Euros/m2 mes)')
grouped_alquiler_medio_sup_mesual_2017.columns = grouped_alquiler_medio_sup_mesual_2017.columns.str.replace('Preu', 'Alquiler_medio(Euros/m2 mes)')
grouped_alquiler_medio_sup_mesual_2016.columns = grouped_alquiler_medio_sup_mesual_2016.columns.str.replace('Preu', 'Alquiler_medio(Euros/m2 mes)')
grouped_alquiler_medio_sup_mesual_2015.columns = grouped_alquiler_medio_sup_mesual_2015.columns.str.replace('Preu', 'Alquiler_medio(Euros/m2 mes)')
```

## 4. Concatenación de los dataframes


## Union de dataframes Alquiler medio por superficie

```python
# relizamos la concatenación de los datasets en un unico dataset
union_alquiler_medio = pd.concat([grouped_alquiler_medio_sup_mesual_2019, grouped_alquiler_medio_sup_mesual_2018, grouped_alquiler_medio_sup_mesual_2017, \
                             grouped_alquiler_medio_sup_mesual_2016, grouped_alquiler_medio_sup_mesual_2015], ignore_index=True)
```
