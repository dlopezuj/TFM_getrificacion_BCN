## 3. Transformación de los datos

```python
# elimino las columnas no necesarias de cada uno de los dataframes
df_ocupacion_2019_drop = df_ocupacion_2019.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri','Poblacio', 'Domicilis'], axis=1)
df_ocupacion_2018_drop = df_ocupacion_2018.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri','Poblacio', 'Domicilis'], axis=1)
df_ocupacion_2017_drop = df_ocupacion_2017.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri','Poblacio', 'Domicilis'], axis=1)
df_ocupacion_2016_drop = df_ocupacion_2016.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Poblacio', 'Domicilis'], axis=1)
df_ocupacion_2015_drop = df_ocupacion_2015.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Poblacio', 'Domicilis'], axis=1)
```

```python
# renombro el nombre de la columna 
df_ocupacion_2019_drop.columns = df_ocupacion_2019_drop.columns.str.replace('Ocupacio_mitjana_(persones_ per_domicili)', 'Ocupacion_media(personas_domicilio)', regex = False)
df_ocupacion_2018_drop.columns = df_ocupacion_2018_drop.columns.str.replace('Ocupacio_mitjana_(persones_ per_domicili)', 'Ocupacion_media(personas_domicilio)', regex = False)
df_ocupacion_2017_drop.columns = df_ocupacion_2017_drop.columns.str.replace('Ocupacio_mitjana_(persones_ per_domicili)', 'Ocupacion_media(personas_domicilio)', regex = False)
df_ocupacion_2016_drop.columns = df_ocupacion_2016_drop.columns.str.replace('Ocupacio_mitjana_(persones_ per_domicili)', 'Ocupacion_media(personas_domicilio)', regex = False)
df_ocupacion_2015_drop.columns = df_ocupacion_2015_drop.columns.str.replace('Ocupacio_mitjana_(persones_ per_domicili)', 'Ocupacion_media(personas_domicilio)', regex = False)
```

```python
# revisamos los resultados
df_ocupacion_2015_drop
```

```python
# en el año 2015, encontramos una categoria de barrio definidad como 99, por lo que lo eliminamos de los datos
df_ocupacion_2015_drop = df_ocupacion_2015_drop[(df_ocupacion_2015_drop['Codi_Barri'] != 99)]
```

## 4. Concatenación de los dataframes

```python
union_ocupacion = pd.concat([df_ocupacion_2019_drop, df_ocupacion_2018_drop, df_ocupacion_2017_drop, \
                             df_ocupacion_2016_drop, df_ocupacion_2015_drop], ignore_index=True)
```
