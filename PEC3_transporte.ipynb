## 3. Transformación de los datos

```python
# elimino las filas con Na en el campo de los ID de cada barrios
df_transporte_filtered_na = df_transporte[df_transporte['addresses_neighborhood_id'].notna()]
```

```python
# realizao un filtrado por los carriles bici, puntos de anclaje y los puntos de Bicing
df_transporte_filtered = df_transporte_filtered_na.loc[(df_transporte_filtered_na['name'].str.contains('Carril ')) \
                                                       | (df_transporte_filtered_na['name'].str.contains('Ancoratge')) \
                                                       #| (df_transporte_filtered_na['name'].str.contains('Parada de taxis')) \
                                                       #| (df_transporte_filtered_na['name'].str.contains("Estació d'Autobusos")) \
                                                       | (df_transporte_filtered_na['name'].str.contains('Bicing'))]
```

```python
df_transporte_filtered.head()
```

```python
df_transporte_filtered.shape
```

```python
# remove all columns except some
df_transporte_filtered_droped = df_transporte_filtered.loc[:, df_transporte_filtered.columns.intersection(['created','modified', 'addresses_neighborhood_id'])]
#convierto a entero la columna del id de los barrios
df_transporte_filtered_droped['addresses_neighborhood_id'] = df_transporte_filtered_droped['addresses_neighborhood_id'].astype(np.int64)
df_transporte_filtered_droped
```

```python
df_transporte_filtered_droped.isna().sum()
```

```python
df_transporte_filtered_droped.describe()
# vemos que el valor máximo y mínimo son 73 y 1, coincide con el número de barrios de barcelona
```

```python
# genero dos nuevas columnas datetime a raiz de las columnas created y modified
df_transporte_filtered_droped['created_new']= pd.to_datetime(df_transporte_filtered_droped['created'],utc=True)
df_transporte_filtered_droped['modified_new']= pd.to_datetime(df_transporte_filtered_droped['modified'],utc=True)
```

```python
df_transporte_filtered_droped
```

```python
# Creamos 5 conjunto de datos filtrando por las instalaciones referentes al transporte que fueron creadas antes de cada año, para obtener cuantos tenemos para cada año
df_transporte_2019 = df_transporte_filtered_droped[(df_transporte_filtered_droped['created_new'] < '2019-01-01')]
df_transporte_2018 = df_transporte_filtered_droped[(df_transporte_filtered_droped['created_new'] < '2018-01-01')]
df_transporte_2017 = df_transporte_filtered_droped[(df_transporte_filtered_droped['created_new'] < '2017-01-01')]
df_transporte_2016 = df_transporte_filtered_droped[(df_transporte_filtered_droped['created_new'] < '2016-01-01')]
df_transporte_2015 = df_transporte_filtered_droped[(df_transporte_filtered_droped['created_new'] < '2015-01-01')]
```

```python
# agrupo por barrio y cuento el número de ubicaciones por barrio
grouped_df_transporte_2019 = df_transporte_2019.groupby(['addresses_neighborhood_id']).count().reset_index()
grouped_df_transporte_2018 = df_transporte_2018.groupby(['addresses_neighborhood_id']).count().reset_index()
grouped_df_transporte_2017 = df_transporte_2017.groupby(['addresses_neighborhood_id']).count().reset_index()
grouped_df_transporte_2016 = df_transporte_2016.groupby(['addresses_neighborhood_id']).count().reset_index()
grouped_df_transporte_2015 = df_transporte_2015.groupby(['addresses_neighborhood_id']).count().reset_index()
```

```python
grouped_df_transporte_2015.head()
```

```python
# eliminamos las columnas que no nos hacen falta
grouped_df_transporte_2019_clean = grouped_df_transporte_2019.drop(['created', 'modified', 'modified_new'], axis=1)
grouped_df_transporte_2018_clean = grouped_df_transporte_2018.drop(['created', 'modified', 'modified_new'], axis=1)
grouped_df_transporte_2017_clean = grouped_df_transporte_2017.drop(['created', 'modified', 'modified_new'], axis=1)
grouped_df_transporte_2016_clean = grouped_df_transporte_2016.drop(['created', 'modified', 'modified_new'], axis=1)
grouped_df_transporte_2015_clean = grouped_df_transporte_2015.drop(['created', 'modified', 'modified_new'], axis=1)
```

```python
# como hay barrios que no estan representados, genero un dataframe de 73 filas con ceros para tener valores de cada barrrio
df = pd.DataFrame(np.arange(1,74), columns=["addresses_neighborhood_id"])
```

```python
# realizo el merge de los datos con el dataframe vacio 
merged_df_transporte_2019_clean = df.merge(grouped_df_transporte_2019_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
merged_df_transporte_2018_clean = df.merge(grouped_df_transporte_2018_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
merged_df_transporte_2017_clean = df.merge(grouped_df_transporte_2017_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
merged_df_transporte_2016_clean = df.merge(grouped_df_transporte_2016_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
merged_df_transporte_2015_clean = df.merge(grouped_df_transporte_2015_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
```

```python
# creamos una nueva columna donde se indica el año de los datos
merged_df_transporte_2019_clean['Any'] = '2019'
merged_df_transporte_2018_clean['Any'] = '2018'
merged_df_transporte_2017_clean['Any'] = '2017'
merged_df_transporte_2016_clean['Any'] = '2016'
merged_df_transporte_2015_clean['Any'] = '2015'
```

```python
# visualizamos los resultados
merged_df_transporte_2019_clean
```

```python
merged_df_transporte_2019_clean.columns = merged_df_transporte_2019_clean.columns.str.replace('created_new', 'transporte')
merged_df_transporte_2018_clean.columns = merged_df_transporte_2018_clean.columns.str.replace('created_new', 'transporte')
merged_df_transporte_2017_clean.columns = merged_df_transporte_2017_clean.columns.str.replace('created_new', 'transporte')
merged_df_transporte_2016_clean.columns = merged_df_transporte_2016_clean.columns.str.replace('created_new', 'transporte')
merged_df_transporte_2015_clean.columns = merged_df_transporte_2015_clean.columns.str.replace('created_new', 'transporte')
```

```python
merged_df_transporte_2015_clean
```

## 4. Concatenación de los dataframes

```python
# relizamos la concatenación de los datasets en un unico dataset
union_transporte = pd.concat([merged_df_transporte_2019_clean, merged_df_transporte_2018_clean, merged_df_transporte_2017_clean, \
                             merged_df_transporte_2016_clean, merged_df_transporte_2015_clean], ignore_index=True)
```

```python
union_transporte
```

```python
union_transporte.columns = union_transporte.columns.str.replace('addresses_neighborhood_id', 'Codi_Barri')
```
