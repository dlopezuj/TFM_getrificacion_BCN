## 3. Transformación de los datos

```python
# eliminamos las columnas que no son necesarias
df_hoteles = df_hoteles.drop(['register_id', 'institution_id', 'institution_name','addresses_roadtype_id', 'addresses_roadtype_name', 'addresses_road_id', 'addresses_road_name' ,\
                              'addresses_start_street_number', 'addresses_end_street_number', 'addresses_neighborhood_name', \
                              'addresses_district_id', 'addresses_district_name', 'addresses_zip_code', 'addresses_town', 'addresses_main_address', \
                              'addresses_type', 'values_id', 'values_attribute_id', 'values_category', 'values_attribute_name', 'values_value', 'values_outstanding', \
                              'values_description', 'secondary_filters_id','secondary_filters_fullpath', 'secondary_filters_tree', \
                              'secondary_filters_asia_id', 'geo_epgs_25831_x', 'geo_epgs_25831_y', 'geo_epgs_4326_x', 'geo_epgs_4326_y', 'secondary_filters_name' \
                              ], axis=1)
```

```python
# visualizamos los resultados
df_hoteles
```

Al eliminar duplicados, vemos que no existen

```python
df_hoteles = df_hoteles.drop_duplicates()
df_hoteles
```

```python
df_hoteles.info()
```

```python
# creamos nuevas columnas de tipo datetime a raiz de creación y modificación, estas columnas parecen indicar cuando se creó el lugar y cuando se ha hecho una última modificación en el documento
df_hoteles['created_new']= pd.to_datetime(df_hoteles['created'],utc=True)
df_hoteles['modified_new']= pd.to_datetime(df_hoteles['modified'],utc=True)
```

```python
df_hoteles
```

```python
# vemos como las nuevas columnas son del tipo datetime
df_hoteles.info()
```

```python
# Creamos 5 conjunto de datos filtrando por los hoteles que fueron creados antes de cada año, para obtener cuantos tenemos para cada año
df_hoteles_2019 = df_hoteles[(df_hoteles['created_new'] < '2019-01-01')]
df_hoteles_2018 = df_hoteles[(df_hoteles['created_new'] < '2018-01-01')]
df_hoteles_2017 = df_hoteles[(df_hoteles['created_new'] < '2017-01-01')]
df_hoteles_2016 = df_hoteles[(df_hoteles['created_new'] < '2016-01-01')]
df_hoteles_2015 = df_hoteles[(df_hoteles['created_new'] < '2015-01-01')]
```

```python
df_hoteles_2019.describe()
```

```python
# eliminamos las columnas que no nos hacen falta
df_hoteles_2019_clean = df_hoteles_2019.drop(['created', 'modified', 'created_new', 'modified_new'], axis=1)
df_hoteles_2018_clean = df_hoteles_2018.drop(['created', 'modified', 'created_new', 'modified_new'], axis=1)
df_hoteles_2017_clean = df_hoteles_2017.drop(['created', 'modified', 'created_new', 'modified_new'], axis=1)
df_hoteles_2016_clean = df_hoteles_2016.drop(['created', 'modified', 'created_new', 'modified_new'], axis=1)
df_hoteles_2015_clean = df_hoteles_2015.drop(['created', 'modified', 'created_new', 'modified_new'], axis=1)
```

```python
# visualizamos los resultados
df_hoteles_2019_clean
```

```python
# agrupamos por barrio y obtenemos el conteo del número de hoteles por barrio
grouped_df_hoteles_2019_clean = df_hoteles_2019_clean.groupby(['addresses_neighborhood_id']).count().reset_index()
grouped_df_hoteles_2018_clean = df_hoteles_2018_clean.groupby(['addresses_neighborhood_id']).count().reset_index()
grouped_df_hoteles_2017_clean = df_hoteles_2017_clean.groupby(['addresses_neighborhood_id']).count().reset_index()
grouped_df_hoteles_2016_clean = df_hoteles_2016_clean.groupby(['addresses_neighborhood_id']).count().reset_index()
grouped_df_hoteles_2015_clean = df_hoteles_2015_clean.groupby(['addresses_neighborhood_id']).count().reset_index()
```

```python
# como hay barrios que no aparecen y para evitar tener null mas adelante, creo un dataframe con 73 elementos y hago un merge con el dataframe de datos rellenando con 0 los null
```

```python
df = pd.DataFrame(np.arange(1,74), columns=["addresses_neighborhood_id"])
```

```python
# realizo el merge de los datos con el dataframe vacio 
merged_df_hoteles_2019_clean = df.merge(grouped_df_hoteles_2019_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
merged_df_hoteles_2018_clean = df.merge(grouped_df_hoteles_2018_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
merged_df_hoteles_2017_clean = df.merge(grouped_df_hoteles_2017_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
merged_df_hoteles_2016_clean = df.merge(grouped_df_hoteles_2016_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
merged_df_hoteles_2015_clean = df.merge(grouped_df_hoteles_2015_clean, on='addresses_neighborhood_id', how='outer').fillna(0)
```

```python
# creamos una nueva columna donde se indica el año de los datos
merged_df_hoteles_2019_clean['Any'] = 2019
merged_df_hoteles_2018_clean['Any'] = 2018
merged_df_hoteles_2017_clean['Any'] = 2017
merged_df_hoteles_2016_clean['Any'] = 2016
merged_df_hoteles_2015_clean['Any'] = 2015
```

```python
# visualizamos los resultados
merged_df_hoteles_2015_clean
```

## 4. Concatenación de los dataframes

```python
# relizamos la concatenación de los datasets en un unico dataset
union_hoteles = pd.concat([merged_df_hoteles_2019_clean, merged_df_hoteles_2018_clean, merged_df_hoteles_2017_clean, \
                         merged_df_hoteles_2016_clean, merged_df_hoteles_2015_clean], ignore_index=True)
```

```python
union_hoteles
```

```python
# revisamos la presencia de valores null
union_hoteles.isnull().any()
```

```python
# renombramos las columnas para poder hacer la unión más adelante
union_hoteles.columns = union_hoteles.columns.str.replace('name', 'Hoteles')
union_hoteles.columns = union_hoteles.columns.str.replace('addresses_neighborhood_id', 'Codi_Barri')
```
