## 3. Transformación de los datos

```python
# elimino las columnas no necesarias de cada uno de los dataframes
df_natalidad_2019_drop = df_natalidad_2019.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Taxa_mil_hab'], axis=1)
df_natalidad_2018_drop = df_natalidad_2018.drop(['Codi_districte', 'Nom_districte', 'Nom_barri', 'Taxa_mil_hab'], axis=1)
df_natalidad_2017_drop = df_natalidad_2017.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri'], axis=1)
df_natalidad_2016_drop = df_natalidad_2016.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri'], axis=1)
df_natalidad_2015_drop = df_natalidad_2015.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri'], axis=1)
```

```python
# Renombro las columna Codi_barri de los siguientes datasets
df_natalidad_2018_drop.columns = df_natalidad_2018_drop.columns.str.replace('Codi_barri', 'Codi_Barri')
df_natalidad_2017_drop.columns = df_natalidad_2017_drop.columns.str.replace('Codi_barri', 'Codi_Barri')
df_natalidad_2016_drop.columns = df_natalidad_2016_drop.columns.str.replace('Codi_barri', 'Codi_Barri')
df_natalidad_2015_drop.columns = df_natalidad_2015_drop.columns.str.replace('Codi_barri', 'Codi_Barri')
```

```python
df_natalidad_2019_drop
```

```python
# renombro las columnas para poder realizar más adelante la concatenación de los conjuntos
df_natalidad_2019_drop.columns = df_natalidad_2019_drop.columns.str.replace('Nombre', 'Tasa_natalidad_mil_hab')
df_natalidad_2018_drop.columns = df_natalidad_2018_drop.columns.str.replace('Nombre', 'Tasa_natalidad_mil_hab')
df_natalidad_2017_drop.columns = df_natalidad_2017_drop.columns.str.replace('Taxa_mil_hab', 'Tasa_natalidad_mil_hab')
df_natalidad_2016_drop.columns = df_natalidad_2016_drop.columns.str.replace('Taxa_mil_hab', 'Tasa_natalidad_mil_hab')
df_natalidad_2015_drop.columns = df_natalidad_2015_drop.columns.str.replace('Taxa_mil_hab', 'Tasa_natalidad_mil_hab')
```

## 4. Concatenación de los dataframes

```python
# relizamos la concatenación de los datasets en un unico dataset
union_natalidad = pd.concat([df_natalidad_2019_drop, df_natalidad_2018_drop, \
                             df_natalidad_2017_drop, df_natalidad_2016_drop, df_natalidad_2015_drop], ignore_index=True)
```

```python
union_natalidad
```

```python
union_natalidad.isna().sum()
```

```python
# La tasa en 2018, tienen una gran cantidad de decimales, para ser consistente con el resto de columnas, dejo únicamente un decimal
union_natalidad['Tasa_natalidad_mil_hab'] = union_natalidad['Tasa_natalidad_mil_hab'].round(decimals = 1)
```
