## 3. Transformación de los datos

```python
Nivell_academic_categorias_2015 = df_academico_2015.Nivell_academic.unique()
Nivell_academic_categorias_2015
```

```python
# realizamos un filtrado por el campo Nivell_academic para los casos que contienen Estudis universitaris
universitarios_2019 = df_academico_2019[df_academico_2019['Nivell_academic'].str.contains("Estudis universitaris")]
universitarios_2018 = df_academico_2018[df_academico_2018['Nivell_academic'].str.contains("Estudis universitaris")]
universitarios_2017 = df_academico_2017[df_academico_2017['Nivell_academic'].str.contains("Estudis universitaris")]
universitarios_2016 = df_academico_2016[df_academico_2016['Nivell_academic'].str.contains("Estudis universitaris")]
universitarios_2015 = df_academico_2015[df_academico_2015['Nivell_academic'].str.contains("Estudis universitaris")]

universitarios_2019
```

```python
# eliminamos las columnas que no necesitamos
universitarios_2019 = universitarios_2019.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Sexe', 'Nivell_academic'], axis=1)
universitarios_2018 = universitarios_2018.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Sexe', 'Nivell_academic'], axis=1)
universitarios_2017 = universitarios_2017.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Sexe', 'Nivell_academic'], axis=1)
universitarios_2016 = universitarios_2016.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Sexe', 'Nivell_academic'], axis=1)
universitarios_2015 = universitarios_2015.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Sexe', 'Nivell_academic'], axis=1)

# reemplazamos la columna Nombre por Estudios_superiores 
universitarios_2019.columns = universitarios_2019.columns.str.replace('Nombre', 'Estudios_superiores')
universitarios_2018.columns = universitarios_2018.columns.str.replace('Nombre', 'Estudios_superiores')
universitarios_2017.columns = universitarios_2017.columns.str.replace('Nombre', 'Estudios_superiores')
universitarios_2016.columns = universitarios_2016.columns.str.replace('Nombre', 'Estudios_superiores')
universitarios_2015.columns = universitarios_2015.columns.str.replace('Nombre', 'Estudios_superiores')
```

```python
# visualizamos los resultados
universitarios_2015
```

```python
# Los datos los tenemos separados para hombres y mujeres, creamos una única categoría con los dos sexos
grouped_univeritarios_2019 = universitarios_2019.groupby(['Codi_Barri', 'Any']).sum().reset_index()
grouped_univeritarios_2018 = universitarios_2018.groupby(['Codi_Barri', 'Any']).sum().reset_index()
grouped_univeritarios_2017 = universitarios_2017.groupby(['Codi_Barri', 'Any']).sum().reset_index()
grouped_univeritarios_2016 = universitarios_2016.groupby(['Codi_Barri', 'Any']).sum().reset_index()
grouped_univeritarios_2015 = universitarios_2015.groupby(['Codi_Barri', 'Any']).sum().reset_index()


grouped_univeritarios_2015
```

```python
grouped_univeritarios_2015
```

## 4. Concatenación de los dataframes

```python
# relizamos la concatenación de los datasets en un unico dataset
union_universitarios = pd.concat([grouped_univeritarios_2019, grouped_univeritarios_2018, grouped_univeritarios_2017, \
                             grouped_univeritarios_2016, grouped_univeritarios_2015], ignore_index=True)
```
