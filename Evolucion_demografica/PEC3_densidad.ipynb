```python
import pandas as pd 
df_densidad_2019=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/datasets/densidad poblacion/2019_densitat.csv')
df_densidad_2018=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/datasets/densidad poblacion/2018_densitat.csv')
df_densidad_2017=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/datasets/densidad poblacion/2017_densitat.csv')
df_densidad_2016=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/datasets/densidad poblacion/2016_densitat.csv')
df_densidad_2015=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/datasets/densidad poblacion/2015_densitat.csv')

```

```python
df_densidad_2016.head()
```

## 1.2 Exploracion de los datos

```python
df_densidad_drop_2019 = df_densidad_2019.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Població', 'Superfície (ha)','Superfície Residencial (ha)', 'Densitat (hab/ha)'], axis=1)
df_densidad_drop_2018 = df_densidad_2018.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Població', 'Superfície (ha)','Superfície Residencial (ha)', 'Densitat (hab/ha)'], axis=1)
df_densidad_drop_2017 = df_densidad_2017.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Població', 'Superfície (ha)','Superfície Residencial (ha)', 'Densitat (hab/ha)'], axis=1)
df_densidad_drop_2016 = df_densidad_2016.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Població', 'Superfície (ha)', 'Densitat (hab/ha)'], axis=1)
df_densidad_drop_2015 = df_densidad_2015.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Població', 'Superfície (ha)', 'Densitat (hab/ha)'], axis=1)
```

```python
df_densidad_drop_2015
```

# Concat

```python
union_densidad = pd.concat([df_densidad_drop_2019, df_densidad_drop_2018, df_densidad_drop_2017, \
                             df_densidad_drop_2016, df_densidad_drop_2015], ignore_index=True)
```

```python
union_densidad.columns = union_densidad.columns.str.replace('Densitat neta (hab/ha)', 'Densidad neta (hab/ha)')
```

```python
union_densidad
```

```python
union_densidad.to_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/output_evolucion_demografica/union_densidad.csv', index = False)
```
