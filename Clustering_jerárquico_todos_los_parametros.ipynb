# Aplicación del clustering jerárquico para todos los parametros extraidos

```python
# importamos librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
# para crear el dendograma
import scipy.cluster.hierarchy as sch
```

```python
# realizamos la lectura del dataset generado
data_gentrification = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/output/df_union_factores.csv')
```

```python
# visualizamos los resultados
data_gentrification
```

```python
# obtenemos los valores para los años 2019, 2018 y 2015 
data_gentrification_2019 = data_gentrification[(data_gentrification['Any'] == 2019)]
data_gentrification_2015 = data_gentrification[(data_gentrification['Any'] == 2015)]
data_gentrification_2018 = data_gentrification[(data_gentrification['Any'] == 2018)]
```

```python
# visualizamos los resultados
data_gentrification_2019
```

```python
#data_gentrification_2015
#data_gentrification_2015 = data_gentrification_2015.drop(['Codi_Barri', 'Any'], axis=1)
```

```python
data_gentrification_2019.reset_index(drop=True, inplace=True)
data_gentrification_2015.reset_index(drop=True, inplace=True)
```

```python
# obtenemos la variación entre los años 2019 y 2015 y 2019 y 2018
df_incremento_2019_2015 = data_gentrification_2019 - data_gentrification_2015
```

```python
df_incremento_2019_2015_select = df_incremento_2019_2015
```

```python
df_incremento_select_2019 = data_gentrification_2019
```

```python
df_incremento_2019_2015_select
```

```python
# eliminamos las Columnas Codi_Barri y Any, no se tendrán en cuenta en la predicción
df_incremento_2019_2015_select = df_incremento_2019_2015_select.drop(['Codi_Barri', 'Any'], axis=1)
```

```python
# visualizamos el tipo de datos
df_incremento_2019_2015_select.dtypes
```

```python
# reasignamos a la variable df_final
df_final = df_incremento_2019_2015_select


df_final
```

```python
# obtenemos la correlación
df_final.corr(method="pearson").abs()
```

```python
# Create correlation matrix
corr_matrix = df_final.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.50)]

# Drop features 
df_final.drop(to_drop, axis=1, inplace=True)
```

```python
# columnas eliminadas por existir correlación
print(to_drop)
```

```python
# visualizamos los resultados
df_final
```

# Normalizamos los valores

```python
# lo primero que haremos será normalizar los valores, para que todas las columnas se encuentren dentro del mismo rango
```

```python
#scaler
scaler = StandardScaler()
scaler.fit(df_final)
df_incremento_norm = scaler.transform(df_final)
```

```python
# visualizamos los resultados
df_incremento_norm
```

```python
x = df_incremento_norm
```

```python
# ward está definido para obtener la distancia euclidiana
dendograma = sch.dendrogram(sch.linkage(x, method='ward'))
plt.title('Dendrograma')
plt.xlabel('Barrios')
plt.ylabel('Distancia Euclidiana')
plt.show()
```

# Predicción para 3 clusters

```python
# realizamos la predicción para 3 clústers
hc_3 = AgglomerativeClustering(n_clusters = 3,
                             affinity= 'euclidean',
                             linkage= 'ward')
```

```python
y_hc_3 = hc_3.fit_predict(x)
```

```python
# agupación para 3 clusters
y_hc_3
```

# Predicción para 4 clusters

```python
# realizamos la predicción para 4 clústers
hc_4 = AgglomerativeClustering(n_clusters = 4,
                             affinity= 'euclidean',
                             linkage= 'ward')
```

```python
y_hc_4 = hc_4.fit_predict(x)
```

```python
# agupación para 4 clusters
y_hc_4
```

```python
len(y_hc_4)
```

```python
clustering_jerarquico_output = data_gentrification_2019
```

```python
# añadimos las predicciones al dataset
clustering_jerarquico_output['clustering_jerarquico_todos_parametrosk3'] = y_hc_3
clustering_jerarquico_output['clustering_jerarquico_todos_parametros_k4'] = y_hc_4
```
