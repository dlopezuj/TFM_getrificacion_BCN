## 3. Transformación de los datos

```python
# elimino las columnas no necesarias
df_migracion_2019_drop = df_migracion_2019.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Taxa_mil_hab'], axis=1)
df_migracion_2018_drop = df_migracion_2018.drop(['Codi_districte', 'Nom_districte', 'Nom_barri', 'Taxa_mil_hab'], axis=1)
df_migracion_2017_drop = df_migracion_2017.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri'], axis=1)
df_migracion_2016_drop = df_migracion_2016.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri'], axis=1)
df_migracion_2015_drop = df_migracion_2015.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri'], axis=1)
```

```python
# visualizamos los resultados obtenidos
df_migracion_2015_drop
```

```python
# renombro las columnas, ya que hay ligeras diferencias entre los archivos de los distintos años
df_migracion_2019_drop.columns = df_migracion_2019_drop.columns.str.replace('Nombre', 'Tasa_migracion_mil_hab')
df_migracion_2018_drop.columns = df_migracion_2018_drop.columns.str.replace('Nombre', 'Tasa_migracion_mil_hab')
df_migracion_2018_drop.columns = df_migracion_2018_drop.columns.str.replace('Codi_barri', 'Codi_Barri')
df_migracion_2017_drop.columns = df_migracion_2017_drop.columns.str.replace('Taxa_mil_hab', 'Tasa_migracion_mil_hab')
df_migracion_2016_drop.columns = df_migracion_2016_drop.columns.str.replace('Taxa_mil_hab', 'Tasa_migracion_mil_hab')
df_migracion_2015_drop.columns = df_migracion_2015_drop.columns.str.replace('Taxa_mil_hab', 'Tasa_migracion_mil_hab')
```

## 4. Concatenación de los dataframes

```python
union_migracion = pd.concat([df_migracion_2019_drop, df_migracion_2018_drop, df_migracion_2017_drop, \
                             df_migracion_2016_drop, df_migracion_2015_drop], ignore_index=True)
```

```python
union_migracion
```

```python
# revisión de valores null
union_migracion.isnull().sum()
```

```python
# la tasa en 2018, tienen una gran cantidad de decimales, para ser consistente con el resto de columnas, dejo únicamente un decimal
union_migracion['Tasa_migracion_mil_hab'] = union_migracion['Tasa_migracion_mil_hab'].round(decimals = 1)
```
