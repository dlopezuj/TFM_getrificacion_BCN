## 3. Transformación de los datos

```python
# elimino de los datasets las columnas que no hacen falta
df_renta_2019_drop = df_renta_2019.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Seccio_Censal'], axis=1)
df_renta_2018_drop = df_renta_2018.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Seccio_Censal'], axis=1)
df_renta_2017_drop = df_renta_2017.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Seccio_Censal'], axis=1)
df_renta_2016_drop = df_renta_2016.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Seccio_Censal'], axis=1)
df_renta_2015_drop = df_renta_2015.drop(['Codi_Districte', 'Nom_Districte', 'Nom_Barri', 'Seccio_Censal'], axis=1)
```

```python
# calculamos la renta media por barrio, con la media de las secciones censales por barrio
grouped_renta_media_2019 = df_renta_2019_drop.groupby(['Codi_Barri']).mean().reset_index()
grouped_renta_media_2018 = df_renta_2018_drop.groupby(['Codi_Barri']).mean().reset_index()
grouped_renta_media_2017 = df_renta_2017_drop.groupby(['Codi_Barri']).mean().reset_index()
grouped_renta_media_2016 = df_renta_2016_drop.groupby(['Codi_Barri']).mean().reset_index()
grouped_renta_media_2015 = df_renta_2015_drop.groupby(['Codi_Barri']).mean().reset_index()
```

```python
# Como hay un espacio en el el nombre del campo Import Euros, renombro la columan
print(grouped_renta_media_2019.columns)
print(grouped_renta_media_2018.columns)
print(grouped_renta_media_2017.columns)
print(grouped_renta_media_2016.columns)
print(grouped_renta_media_2015.columns)
```

```python
# renombramos la columna
grouped_renta_media_2018.columns = grouped_renta_media_2018.columns.str.replace('Import_Euros ', 'Import_Euros')
grouped_renta_media_2017.columns = grouped_renta_media_2017.columns.str.replace('Import_Euros ', 'Import_Euros')
grouped_renta_media_2016.columns = grouped_renta_media_2016.columns.str.replace('Import_Euros ', 'Import_Euros')
grouped_renta_media_2015.columns = grouped_renta_media_2015.columns.str.replace('Import_Euros ', 'Import_Euros')
```

```python
# visualizamos los resultados
grouped_renta_media_2015
```

```python

# elimino la columna any y renombro la columna valor para poder hacer el join
grouped_renta_media_2019.columns = grouped_renta_media_2019.columns.str.replace('Import_Euros', 'renta_media_hogar(euros)')
grouped_renta_media_2018.columns = grouped_renta_media_2018.columns.str.replace('Import_Euros', 'renta_media_hogar(euros)')
grouped_renta_media_2017.columns = grouped_renta_media_2017.columns.str.replace('Import_Euros', 'renta_media_hogar(euros)')
grouped_renta_media_2016.columns = grouped_renta_media_2016.columns.str.replace('Import_Euros', 'renta_media_hogar(euros)')
grouped_renta_media_2015.columns = grouped_renta_media_2015.columns.str.replace('Import_Euros', 'renta_media_hogar(euros)')
```

## 4. Concatenación de los dataframes

```python
union_renta_media = pd.concat([grouped_renta_media_2019, grouped_renta_media_2018, grouped_renta_media_2017, \
                             grouped_renta_media_2016, grouped_renta_media_2015], ignore_index=True)
```
