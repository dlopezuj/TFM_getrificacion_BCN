Después del análisis observado, vemos que la limpieza de datos que tenemos que realizar está enfocada en la eliminación del dataset de los valores no válidos, como es el caso de los -1, en los campos de distrito y barrio.

```python
selected_indicents = ['AGRESSIONS','ACCIDENTS DE TRÀNSIT AMB FERITS', 'VANDALISME', 'CONFLICTES EN LOCALS', "DELICTES CONTRA L'ORDRE PÚBLIC", "VENDA AMBULANT" , "OCUPACIONS IL·LÍCITES", "INFRACCIONS D'ESTRANGERS" , "TRIBUS URBANES" ]
#selected_indicents = ['AGRESSIONS', 'BARALLES']

selected_indicents
```

```python
# Selecciono las columnas de los incidentes que tiene que ver con la delinquencia
cols = df_incidentes_2015.select_dtypes(object).columns
```

```python
# elimino los caracteres como espacios y tabs de las columnas
df_incidentes_2019[cols] = df_incidentes_2019[cols].apply(lambda x: x.str.strip())
df_incidentes_2018[cols] = df_incidentes_2018[cols].apply(lambda x: x.str.strip())
df_incidentes_2017[cols] = df_incidentes_2017[cols].apply(lambda x: x.str.strip())
df_incidentes_2016[cols] = df_incidentes_2016[cols].apply(lambda x: x.str.strip())
df_incidentes_2015[cols] = df_incidentes_2015[cols].apply(lambda x: x.str.strip())
```

```python
df_incidentes_2015
```

```python
# selecciono del conjunto original de datos los eventos que son relevantes con la criminalidad
incidentes_criminalidad_2019 = df_incidentes_2019[df_incidentes_2019['Descripcio_Incident'].isin(selected_indicents)]
incidentes_criminalidad_2018 = df_incidentes_2018[df_incidentes_2018['Descripcio_Incident'].isin(selected_indicents)]
incidentes_criminalidad_2017 = df_incidentes_2017[df_incidentes_2017['Descripcio_Incident'].isin(selected_indicents)]
incidentes_criminalidad_2016 = df_incidentes_2016[df_incidentes_2016['Descripcio_Incident'].isin(selected_indicents)]
incidentes_criminalidad_2015 = df_incidentes_2015[df_incidentes_2015['Descripcio_Incident'].isin(selected_indicents)]
```

```python
incidentes_criminalidad_2015
```

```python
cleaned_df_incidentes_criminalidad_2019 = incidentes_criminalidad_2019[df_incidentes_2019['Codi_barri'] != -1]
cleaned_df_incidentes_criminalidad_2018 = incidentes_criminalidad_2018[df_incidentes_2018['Codi_barri'] != -1]
cleaned_df_incidentes_criminalidad_2017 = incidentes_criminalidad_2017[df_incidentes_2017['Codi_barri'] != -1]
cleaned_df_incidentes_criminalidad_2016 = incidentes_criminalidad_2016[df_incidentes_2016['Codi_barri'] != -1]
cleaned_df_incidentes_criminalidad_2015 = incidentes_criminalidad_2015[df_incidentes_2015['Codi_barri'] != -1]
```

```python
cleaned_df_incidentes_criminalidad_2015
```

```python
cleaned_df_incidentes_criminalidad_2015.loc[cleaned_df_incidentes_criminalidad_2015['Codi_barri'].idxmin()]
```

```python
# elimino las columnas que no me interesan
incidents19 = cleaned_df_incidentes_criminalidad_2019.drop(['Codi_Incident', 'Descripcio_Incident', 'Codi_districte', 'Mes_any', 'Nom_mes'], axis=1)
incidents18 = cleaned_df_incidentes_criminalidad_2018.drop(['Codi_Incident', 'Descripcio_Incident', 'Codi_districte', 'Mes_any', 'Nom_mes'], axis=1)
incidents17 = cleaned_df_incidentes_criminalidad_2017.drop(['Codi_Incident', 'Descripcio_Incident', 'Codi_districte', 'Mes_any', 'Nom_mes'], axis=1)
incidents16 = cleaned_df_incidentes_criminalidad_2016.drop(['Codi_Incident', 'Descripcio_Incident', 'Codi_districte', 'Mes_any', 'Nom_mes'], axis=1)
incidents15 = cleaned_df_incidentes_criminalidad_2015.drop(['Codi_Incident', 'Descripcio_Incident', 'Codi_districte', 'Mes_any', 'Nom_mes'], axis=1)
```

```python
# agrupo el número de incidentes por Distrito, barrios, mes y año
grouped_indicents_19 = incidents19.groupby(['Codi_barri','Nom_districte', 'Nom_barri', 'NK_Any']).sum().reset_index()
grouped_indicents_18 = incidents18.groupby(['Codi_barri','Nom_districte', 'Nom_barri', 'NK_Any']).sum().reset_index()
grouped_indicents_17 = incidents17.groupby(['Codi_barri','Nom_districte', 'Nom_barri', 'NK_Any']).sum().reset_index()
grouped_indicents_16 = incidents16.groupby(['Codi_barri','Nom_districte', 'Nom_barri', 'NK_Any']).sum().reset_index()
grouped_indicents_15 = incidents15.groupby(['Codi_barri','Nom_districte', 'Nom_barri', 'NK_Any']).sum().reset_index()
```

```python
grouped_indicents_15
```

```python
grouped_indicents_19
```

```python
grouped_indicents_19.columns = grouped_indicents_19.columns.str.replace('Numero_incidents_GUB', 'incidentes')
grouped_indicents_18.columns = grouped_indicents_18.columns.str.replace('Numero_incidents_GUB', 'incidentes')
grouped_indicents_17.columns = grouped_indicents_17.columns.str.replace('Numero_incidents_GUB', 'incidentes')
grouped_indicents_16.columns = grouped_indicents_16.columns.str.replace('Numero_incidents_GUB', 'incidentes')
grouped_indicents_15.columns = grouped_indicents_15.columns.str.replace('Numero_incidents_GUB', 'incidentes')

grouped_indicents_19.columns = grouped_indicents_19.columns.str.replace('NK_Any', 'Any')
grouped_indicents_18.columns = grouped_indicents_18.columns.str.replace('NK_Any', 'Any')
grouped_indicents_17.columns = grouped_indicents_17.columns.str.replace('NK_Any', 'Any')
grouped_indicents_16.columns = grouped_indicents_16.columns.str.replace('NK_Any', 'Any')
grouped_indicents_15.columns = grouped_indicents_15.columns.str.replace('NK_Any', 'Any')
```

```python
grouped_indicents_19 = grouped_indicents_19.drop(['Nom_districte','Nom_barri'], axis=1)
grouped_indicents_18 = grouped_indicents_18.drop(['Nom_districte','Nom_barri'], axis=1)
grouped_indicents_17 = grouped_indicents_17.drop(['Nom_districte','Nom_barri'], axis=1)
grouped_indicents_16 = grouped_indicents_16.drop(['Nom_districte','Nom_barri'], axis=1)
grouped_indicents_15 = grouped_indicents_15.drop(['Nom_districte','Nom_barri'], axis=1)
```

```python
grouped_indicents_16
```

## 4. Concatenación de los dataframes

```python
# relizamos la concatenación de los datasets en un unico dataset
union_incidentes = pd.concat([grouped_indicents_19, grouped_indicents_18, grouped_indicents_17, \
                             grouped_indicents_16, grouped_indicents_15], ignore_index=True)
```

```python
union_incidentes
```

```python
# renombramos Codi barri
union_incidentes.columns = union_incidentes.columns.str.replace('Codi_barri', 'Codi_Barri')
```
