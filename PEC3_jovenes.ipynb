## 3. Transformación de los datos

```python
# convierto la columna de edad a númerico
df_edades_2019['Edat_int'] = df_edades_2019['Edat_any_a_any'].str.split(' ').str[0].astype(int)
df_edades_2018['Edat_int'] = df_edades_2018['Edat_any_a_any'].str.split(' ').str[0].astype(int)
df_edades_2017['Edat_int'] = df_edades_2017['Edat_any_a_any'].str.split(' ').str[0].astype(int)
df_edades_2016['Edat_int'] = df_edades_2016['Edat_any_a_any'].str.split(' ').str[0].astype(int)
df_edades_2015['Edat_int'] = df_edades_2015['Edat_any_a_any'].str.split(' ').str[0].astype(int)
```

Obtengo la cantidad de Jóvenes filtrando el número de personas comprendidades entre los 25 y los 35 años

```python
jovenes_2019 = df_edades_2019[(df_edades_2019['Edat_int'] >= 25) & (df_edades_2019['Edat_int'] <= 35)]
jovenes_2018 = df_edades_2018[(df_edades_2018['Edat_int'] >= 25) & (df_edades_2018['Edat_int'] <= 35)]
jovenes_2017 = df_edades_2017[(df_edades_2017['Edat_int'] >= 25) & (df_edades_2017['Edat_int'] <= 35)]
jovenes_2016 = df_edades_2016[(df_edades_2016['Edat_int'] >= 25) & (df_edades_2016['Edat_int'] <= 35)]
jovenes_2015 = df_edades_2015[(df_edades_2015['Edat_int'] >= 25) & (df_edades_2015['Edat_int'] <= 35)]

jovenes_2019
```

```python
# elimino las columnas que no son necesarias
jovenes_2019 = jovenes_2019.drop(['Codi_Districte','Nom_Districte', 'Nom_Barri', 'Sexe', 'Edat_int', 'Edat_any_a_any'], axis=1)
jovenes_2018 = jovenes_2018.drop(['Codi_Districte','Nom_Districte', 'Nom_Barri', 'Sexe', 'Edat_int', 'Edat_any_a_any'], axis=1)
jovenes_2017 = jovenes_2017.drop(['Codi_Districte','Nom_Districte', 'Nom_Barri', 'Sexe', 'Edat_int', 'Edat_any_a_any'], axis=1)
jovenes_2016 = jovenes_2016.drop(['Codi_Districte','Nom_Districte', 'Nom_Barri', 'Sexe', 'Edat_int', 'Edat_any_a_any'], axis=1)
jovenes_2015 = jovenes_2015.drop(['Codi_Districte','Nom_Districte', 'Nom_Barri', 'Sexe', 'Edat_int', 'Edat_any_a_any'], axis=1)
```

```python
# visualizamos los resultados obtenidos
jovenes_2019
```

```python
# se realiza la agrupación por barrio para los datasets procesados y realizamos una visualización del resultado para realizar una verifiación
grouped_jovenes_2019 = jovenes_2019.groupby(['Codi_Barri', 'Any']).sum().reset_index()
grouped_jovenes_2018 = jovenes_2018.groupby(['Codi_Barri', 'Any']).sum().reset_index()
grouped_jovenes_2017 = jovenes_2017.groupby(['Codi_Barri', 'Any']).sum().reset_index()
grouped_jovenes_2016 = jovenes_2016.groupby(['Codi_Barri', 'Any']).sum().reset_index()
grouped_jovenes_2015 = jovenes_2015.groupby(['Codi_Barri', 'Any']).sum().reset_index()
grouped_jovenes_2019
```

```python
# se renombra la columna nombre con el valor jovenes
grouped_jovenes_2019.columns = grouped_jovenes_2019.columns.str.replace('Nombre', 'jovenes')
grouped_jovenes_2018.columns = grouped_jovenes_2018.columns.str.replace('Nombre', 'jovenes')
grouped_jovenes_2017.columns = grouped_jovenes_2017.columns.str.replace('Nombre', 'jovenes')
grouped_jovenes_2016.columns = grouped_jovenes_2016.columns.str.replace('Nombre', 'jovenes')
grouped_jovenes_2015.columns = grouped_jovenes_2015.columns.str.replace('Nombre', 'jovenes')
```

```python
# en el año 2015 tenemos un codigo de barrio como 99, por lo que procedemos a eliminarlo
grouped_jovenes_2015 = grouped_jovenes_2015[(grouped_jovenes_2015['Codi_Barri'] != 99)]
```

```python
grouped_jovenes_2015
```

## 4. Concatenación de los dataframes

```python
# relizamos la concatenación de los datasets en un unico dataset
union_jovenes = pd.concat([grouped_jovenes_2019, grouped_jovenes_2018, grouped_jovenes_2017, \
                             grouped_jovenes_2016, grouped_jovenes_2015], ignore_index=True)
```
