# Realizamos la unión de los dataframes

```python
# importamos librerías
import pandas as pd 
```

```python
from google.colab import drive 
drive.mount('/content/gdrive')
```

```python
# realizamos la lectura de todos los archivos
union_natalidad = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/output_evolucion_demografica/union_natalidad.csv")
union_jovenes = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/output_evolucion_demografica/union_jovenes.csv")
union_ocupacion_media = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/output_evolucion_demografica/union_ocupacion.csv")
union_migracion = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/output_evolucion_demografica/union_migracion.csv")
union_densidad = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_demografica/output_evolucion_demografica/union_densidad.csv')
#
union_renta_media = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Evolucion_economica/output_evolucion_economica/union_renta_media.csv")
union_alquiler_medio = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Evolucion_economica/output_evolucion_economica/union_alquiler_medio.csv")
union_precio_medio = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Evolucion_economica/output_evolucion_economica/union_precio_medio.csv")
#
union_transporte = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_servicios/output_evolucion_servicios/union_transporte.csv')
union_cultura = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_servicios/output_evolucion_servicios/union_cultura.csv')
union_restaurantes = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_servicios/output_evolucion_servicios/union_restaurantes.csv')
union_parques = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_servicios/output_evolucion_servicios/union_parques.csv')
union_hoteles = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_servicios/output_evolucion_servicios/union_hoteles.csv')
#
union_criminalidad = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_social/output_evolucion_social/union_incidentes.csv')
union_universitarios = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Evolucion_social/output_evolucion_social/union_universitarios.csv')
```

Join de los Dataframes

```python
df_union_all = union_natalidad.set_index(['Codi_Barri','Any']) \
.join(union_jovenes.set_index(['Codi_Barri','Any'])) \
.join(union_ocupacion_media.set_index(['Codi_Barri','Any'])) \
.join(union_migracion.set_index(['Codi_Barri','Any'])) \
.join(union_densidad.set_index(['Codi_Barri','Any'])) \
.join(union_renta_media.set_index(['Codi_Barri','Any'])) \
.join(union_alquiler_medio.set_index(['Codi_Barri','Any'])) \
.join(union_precio_medio.set_index(['Codi_Barri','Any'])) \
.join(union_transporte.set_index(['Codi_Barri','Any'])) \
.join(union_cultura.set_index(['Codi_Barri','Any'])) \
.join(union_restaurantes.set_index(['Codi_Barri','Any'])) \
.join(union_parques.set_index(['Codi_Barri','Any'])) \
.join(union_hoteles.set_index(['Codi_Barri','Any'])) \
.join(union_criminalidad.set_index(['Codi_Barri','Any'])) \
.join(union_universitarios.set_index(['Codi_Barri','Any'])) \
.sort_values(by=['Codi_Barri','Any']).reset_index()
df_union_all
```
